flights.csv - a list of flight details for various airlines in 2015




Method 1 -

Loading using the SQL query method




```python
myconn = duckdb.connect(database='mydb.thanga')
```

```python
myconn.execute(
    '''
CREATE TABLE flights
as 
SELECT
*
FROM
read_csv_auto('flights.csv')
'''
)
```

```python
myconn.execute(
    '''
select * from flights'''
).df()
```



Or we can also use



```python
myconn.execute(
    '''
create table flights_copy as from 'flights.csv' '''
).df()
```
